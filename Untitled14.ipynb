-> Objective:
To create a model to forecast the demand or the passenger traffic in airplanes using time series analysis.

-> Time Series Analysis:

It refers to a line or series of observations that are recorded at equal intervals.
Analysis is possible which could predict future values based on prior observed values.
Only two variables, viz., time and the variable to be forecasted.

-> Why and Where Time Series is Used:

Business forecasting
Understanding past behavior
Planning the future
Evaluating current accomplishments
Importance in economics, finance, healthcare, environmental science, and sales forecasting.
Components of Time Series:

-> Trend: an upward or downward movement in time.
Seasonality: seasonal variances.
Noise or Irregularity: random spikes and troughs.
Cyclicity: behavior repeating after large intervals.
Stationarity:

A time series must be stationary before trying to fit statistical models on them.
Constant mean, variance, and auto-covariance over the time series.
Trend and seasonality make it non-stationary.


-> Why Stationarity Matters:

Stationary data means that its mean and standard deviation do not change over time.
ARIMA requires stationary data.


-> Tests for Stationarity

a. Rolling Statistics: plot moving average or standard deviation in order to detect change over time.
b. Augmented Dickey-Fuller test: helps with detection of stationarity, and returns a p-value

-> ARIMA Model:
Mixture of AR and MA models.
Three hyperparameters: P (auto-regressive lags), d (order of differentiation), and Q (moving average).
AR part: correlation between preceding and present time periods.
MA part: smoothes noise
I part: links AR and MA parts together

-> Determining P and Q for ARIMA:

The order of P and Q can be obtained using plots of ACF and PACF.
The x-value where the graph line crosses the y-axis for the first time is noted.

-> Exponential Smoothing:
Smoothing technique in rule of thumb for smoothing time series data
Accords exponentially decreasing weights over time
It is normally used to analyze the time-series data.

Let's make one function consisting of stationary data checking and ADCF test working. Because we will need to repeat the steps many times, therefore, making function will become very handy

```python
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
import pandas as pd

def test_stationarity(timeseries):
    # Determine rolling statistics
    movingAverage = timeseries.rolling(window=12).mean()
    movingSTD = timeseries.rolling(window=12).std()

    # Plot rolling statistics
    fig, ax = plt.subplots(2, 1, figsize=(10, 6))

    ax[0].plot(timeseries, color='blue', label='Original')
    ax[0].plot(movingAverage, color='red', label='Rolling Mean')
    ax[0].legend(loc='best')
    ax[0].set_title('Rolling Mean')
    ax[0].set_xlabel('Time')
    ax[0].set_ylabel('Passengers')

    ax[1].plot(movingSTD, color='black', label='Rolling Std')
    ax[1].legend(loc='best')
    ax[1].set_title('Rolling Standard Deviation')
    ax[1].set_xlabel('Time')
    ax[1].set_ylabel('Passengers')

    plt.tight_layout()
    plt.show(block=False)

    # Perform DickeyFuller test:
    print('Results of Dickey Fuller Test:')
    df_test = adfuller(timeseries['Passengers'], autolag='AIC')
    dfoutput = pd.Series(df_test[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in df_test[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
test_stationarity(dfind)
```

From above plot, we can see that Rolling Mean itself has a trend component even though Rolling Standard Deviation is fairly constant with time.

For time series to be stationary, we need to ensure that both Rolling Mean and Rolling Standard Deviation remain fairly constant WRT time.

Both the curves needs to be parallel to X-Axis, in our case it is not so.

```python
df_log = np.log(dfind)
```

```python
import matplotlib.pyplot as plt

# Plot the log-transformed data
plt.figure(figsize=(10, 6))
plt.plot(df_log, marker='o', linestyle='-', color='blue', alpha=0.7)
plt.title('Log-Transformed Data')
plt.xlabel('Time')
plt.ylabel('Log(Passengers)')
plt.grid(True)
plt.show(block=False)
```

```python
rollmean_log = df_log.rolling(window=12).mean()
rollstd_log = df_log.rolling(window=12).std()
```

```python
plt.plot(df_log, color='blue', label='Original')
plt.plot(rollmean_log, color='red', label='Rolling Mean')
plt.plot(rollstd_log, color='black', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation (Logarithmic Scale)')
```

From above graph we can say that, we slightly bettered our previous results. Now, we are heading into the right direction.

From the above graph, Time series with log scale as well as Rolling Mean(moving avg) both have the trend component. Thus subtracting one from the other should remove the trend component.

R (result) = Time Series Loca Scale - Rolling Mean Log Scale -> this can be our final non trend curve

```python
df_new = df_log - rollmean_log
```

```python
df_new.head()
```

```python
df_new.dropna(inplace=True)
```

```python
df_new.head()
```

```python
test_stationarity(df_new)
```

From the above plot, we came to know that "indeed subtracting two related series having similar trend components actually removed trend and made the dataset stationary"

Also, after concluding the results from ADFC test:

p-value has reduced from 0.99 to 0.022
Critical values at 1%,5%,10% confidence intervals are pretty close to the Test Statistic
So we can now say that given series is now STATIONARY
